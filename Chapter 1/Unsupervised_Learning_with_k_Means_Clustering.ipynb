# Unsupervised Learning

*   purpose :
  1. to provide insights into existing data
  2. to group data into categories and categorize future inputs accordingly
*   don't require labeled data

# Clustering

* purpose : to group data by similarity
* most popular clustering algorithm : **k-means clustering**
  * takes *n* data smaples and groups them into *m* clusters, where *m* is a number you specify

# Grouping
* performed using an iterative process that computes a **centroid** for each cluster
* assigns samples to clusters based on their proximity to the cluster centroid
  *  the sample is allocated to the cluster with the shortest distance between that sample and the centroid of the cluster


```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
from sklearn.datasets import make_blobs

points, cluster_indexes = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=0)

x=points[:, 0]
y=points[:, 1]

plt.scatter(x, y, s=50, alpha=0.7)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4, random_state=0)
kmeans.fit(points)
predicted_cluster_indexes=kmeans.predict(points)

plt.scatter(x, y, c=predicted_cluster_indexes, s=50, alpha=0.7, cmap='viridis')

centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="red", s=100)
```

```python
inertias = []


#i => number of clusters
for i in range(1,10):
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(points)
  inertias.append(kmeans.inertia_)

plt.plot(range(1,10), inertias)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
```
